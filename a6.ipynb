Conceitos importantes relacionados aos grafos.

- Grau de um vértice;
- Caminho entre dois vértices;
- Distância entre dois vértices;
- Componente conexa;
- Densidade de um grafo;
- Diâmetro de um grafo é o maior entre os menores caminhos de um vértice a outro;
- Caminho médio entre um vértice e outro
- Densidade de um grafo avalia a quantidade de arestas: um grafo de densidade máxima (1) é um grafo completo, pois possui todas as arestas.

```python
# Calculando grau médio
# Quanto graus tem em média um vértice do meu grafo

import networkx as nx
import matplotlib.pyplot as plt

g = nx.Graph()
g.add_node("A", tipo="letra", cor="azul")
g.add_node("B", tipo="letra")
g.add_node("3", tipo="numero")
g.add_node("4", tipo="numero")
g.add_edge("A", "3", peso=5)
g.add_edge("B", "3", peso=3)
g.add_edge("3", "4", peso=2)

nx.draw(g, with_labels=True, node_size=800)
nx.spring_layout(g)
plt.show()

# Função que retorna uma tupla com o vértice e o grau do vértice
print(g.degree())

graus = g.degree()
# Obtém a soma dos graus - poderia ter usado ao inves de graus, o g.nodes():

# grau medio = sum(grau_cada_vertice)/g.number_of_nodes()

acumulador = 0
for vertice in g.nodes():
    acumulador += graus[vertice]
    print(graus[vertice])
print(acumulador)

grau_medio = acumulador / g.number_of_nodes()
print(grau_medio)
```
