Grafos bipartidos

```python
import networkx as nx
import matplotlib.pyplot as plt

# v√©rtices
livros = ['L1', 'L2', 'L3', 'L4']
alunos = ['A1', 'A2', 'A3']

emprestimos = [("L1", "A2"), ("L1", "A3"), ('L2', 'A1'), ('L3', 'A1'), ('L3', 'A2'), ('L4', 'A1')]

grafo = nx.Graph()
grafo.add_nodes_from(livros)
grafo.add_nodes_from(alunos)
grafo.add_edges_from(emprestimos)

# define os vertices do lado esquerdo
esquerda = livros
# define o layout do grafo, passando ele e o como alinhar
posicao = nx.bipartite_layout(grafo, esquerda)

nx.draw_networkx(grafo, pos=posicao)

print(grafo.nodes())
```
