A Centralidade de Grau vai mostrar a probabilidade de um vértice X estar conectado ao vértice Y

A Centralidade de grau é dada por

C(v) = grau(v) / n - 1

```python
import networkx as nx
import matplotlib.pyplot as plt

g2 = nx.Graph()
g2.add_edge('A', 'B')
g2.add_edge('A', 'C')
g2.add_edge('B', 'C')
g2.add_edge('B', 'D')

nx.draw(g2, with_labels=True)
plt.show()


for i in g2.nodes():
    calculoCentralidadeDeGrau = g2.degree(i) / (g2.number_of_nodes() - 1)
    print(f"{i}: {calculoCentralidadeDeGrau:.2f}")

lista_centralidade_de_grau = nx.degree_centrality(g2)
print(lista_centralidade_de_grau)
```

-  Centralidade de autovetor: esse tipo de centralidade análise um vértice e os vértices que se relacionam com ele, mais precisamente a influência que os vértices adjacentes possuem em relação ao "vértice origem". (PageRank).

```python
lista_centralidade_autovetor = nx.eigenvector_centrality(g2)
print(str(lista_centralidade_autovetor))
```
