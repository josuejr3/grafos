Centralidade de Proximidade em Grafos

- A centralidade de proximidade indica quem é o "cara" que está mais próximo de um determinado grupo.

- Quem tem a menor distância para os outros

```python
import networkx as nx
import matplotlib.pyplot as plt

g2 = nx.Graph()
g2.add_edge('A', 'B')
g2.add_edge('A', 'C')
g2.add_edge('B', 'C')
g2.add_edge('B', 'D')
nx.draw_networkx(g2, node_color='#008080', with_labels=True, font_color='white')
plt.show()
print(g2)
```

A fórmula da centralidade de proximidade de um grafo é dada pela quantidade de vértices - 1 dividida pela soma das distâncias do vértice em análise em relação aos outros.

Obs: o -1 entra no denominador também pois exclui-se o vértice "fonte'

O valor máximo é 1 e o mínimo é 0, quanto mais proximo dos outros vértices mais próximo de 1

```python
print(nx.closeness_centrality(g2))
```
